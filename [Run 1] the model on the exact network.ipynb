# SIMULATION (NODE-BASED MODEL)

```python
from model_v3 import NodeBasedModel
```

Below we upload back the processed networks and train delay data.

```python
year = 2019
month = 4

graph_list_path = f"./networks_for_node_model_{month:02d}-{year}.pkl"

node_based = NodeBasedModel(graph_list_path)

# STORE THE RAW DATA
node_based.store_data_from_directory(processed_json_data_dir)
```

Choose the day of the week for which to make simultaion.

```python
# CHOOSE THE DAY OF THE WEEK AND THE DATA FILE PATH AND USE THE DATA
day_of_week = 3
data_path = node_based.weekday_dict[day_of_week][1]
node_based.upload_raw_data(data_path)

# CALCULATE MODEL PARAMETERS
g = node_based.graphs_list[day_of_week]
node_based.calculate_model_parameters(g)
```

## Start

Now we can start the simulation. Choose the resolution parameter for discrete simulation. Note that very small numbers may affect the convergence of the algorithm. Good practice is to use 30 secs timeframe `dt`.

```python
# GLOBAL RESOLUTION PARAMETER (IN SECONDS)
dt = 30

node_based.parse_raw_data(dt)
```

```python
# SIMULATION OF THE MODEL
from datetime import datetime, timedelta
from model_v3.utils import get_period_of_day

from model_v3.simulation import simulate_model


T_duration = 1*3600      # how long to simulate in seconds 
dt_simulation = 30       # simulation resolution
start_simulation_time = node_based.start_day_dt + timedelta(hours = 17, minutes = 10) # Time to start simulation
end_simulation_time = start_simulation_time + timedelta(seconds = T_duration) # Time to stop simulation

start_t = int((start_simulation_time - node_based.start_day_dt).total_seconds()/dt_simulation)
end_t = int(((start_simulation_time - node_based.start_day_dt).total_seconds() + T_duration)/dt_simulation)


NUM_PERIODS = g.graph["num_periods"]
dp = get_period_of_day(NUM_PERIODS) # dictionary to map hours to day periods
simulate_hour = start_simulation_time.hour
index_hour = dp[simulate_hour]

G = node_based.G[index_hour]
B_vector = node_based.B_vector[index_hour]
initial_delay_vec = node_based.real_node_ts[start_t, :]


tt, X_dt, node_based_sim_delay_ts = simulate_model(g, G, B_vector, start_simulation_time, 
                                                        T_duration, dt_simulation, initial_delay_vec,
                                                        verbose = False)
```

## Plot

Plot the time series of delays at random set of L points during the simulation.

```python
import matplotlib.pyplot as plt
import numpy as np
N = len(g)
L = 30
sample_points = np.random.randint(0, N, size = L)

fig = plt.figure(figsize = (15,4), dpi = 90)
for i in sample_points:# increasing_delay_indices:
    plt.plot(X_dt, node_based_sim_delay_ts[:,i], lw = 1)
plt.xlabel("Time")
plt.ylabel("Delay")
plt.show()
```

```python
# PLOT THE EVOLUTION OF THE TOTAL DELAY IN THE DATA
import numpy as np
import matplotlib.pyplot as plt

real_total_delay = np.sum(node_based.real_node_ts_smooth, axis = 1)
num_timestamps = real_total_delay.shape[0]
X = [node_based.start_day_dt + timedelta(seconds= dt*i) for i in range(num_timestamps)]

plt.figure(figsize = (16,2), dpi = 90)
plt.plot(X, real_total_delay, lw = 1.5, label = "Total delay in the system")
plt.axvspan(start_simulation_time, end_simulation_time, lw = 2, ls = "--", color = "red", 
            alpha = 0.3, label = "Simulation part")
plt.legend(loc = "best")
plt.xlabel("Time")
plt.show()
```

```python
# PLOT THE EVOLUTION OF THE TOTAL DELAY SIMULATION
import matplotlib.pyplot as plt

real_total_delay = np.sum(node_based_sim_delay_ts, axis = 1)

plt.figure(figsize = (16,2), dpi = 90)
plt.plot(X_dt, real_total_delay, lw = 0.5, label = "reality")
plt.axvline(start_simulation_time, lw = 2, ls = "--", color = "red")
plt.axvline(end_simulation_time, lw = 2, ls = "--", color = "red")
plt.show()
```

```python
# PLOT THE EVOLUTION OF THE TOTAL DELAY IN THE REAL SYSTEM
real_total_delay = np.sum(node_based.real_node_ts[start_t:end_t, :], axis = 1)

plt.figure(figsize = (16,2), dpi = 90)
plt.plot(X_dt, real_total_delay, lw = 1.5,)
plt.show()
```
